```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
```

```python
def get_data(filename):
    data = pd.read_csv(filename, header=None)
    #data = data.head(5)
    data = data.values
    return(data)
```

```python
data = get_data('entropy_comparison_3.csv')

x = data[0,:]
original = data[1,:]
randomized = data[2,:]
ecl = data[3,:]
js = data[4,:]
```

```python
plt.figure(figsize=(15,10))
plt.plot(x, original)
plt.plot(x, randomized)
plt.plot(x, ecl)
plt.plot(x, js)
plt.show()
```
